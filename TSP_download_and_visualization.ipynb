```python
gene_names = adata.var.index.tolist()
expression_matrix = adata.X.toarray()
```

```python
from sklearn.preprocessing import normalize
import pandas as pd

normalized_expression_matrix = normalize(expression_matrix, norm='l1', axis=1)
normalized_df = pd.DataFrame(normalized_expression_matrix, columns=gene_names)
```

```python
import numpy as np
documents = []
for cell_idx in range(expression_matrix.shape[0]):
    expressed_genes = [gene_names[i] for i in np.nonzero(expression_matrix[cell_idx, :])[0]]
    documents.append(" ".join(expressed_genes))
```

```python
print (normalized_df.iloc[4].max())
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of the expression values
plt.hist(normalized_df.values.flatten(), bins=100)
plt.xlabel('Normalized Expression Values')
plt.ylabel('Frequency')
plt.title('Distribution of Normalized Gene Expression Values')
plt.show()
```

```python
import numpy as np

# Flatten the normalized expression values to a 1D array
expression_values = normalized_df.values.flatten()

# Calculate quantiles
quantiles = np.percentile(expression_values, [90, 95, 99, 99.9])
print("Quantiles (90th, 95th, 99th, 99.9th):", quantiles)

# Choose a threshold based on these quantiles
threshold = quantiles[2]  # For example, using the 99th percentile
print("Chosen threshold:", threshold)
```

```python
mask = normalized_df > threshold

# Use the mask to create documents
documents = (mask.apply(lambda row: ' '.join(row.index[row]), axis=1)
                .replace('', 'no_significant_genes')
                .tolist())
```

```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("ctheodoris/Geneformer")
model = AutoModel.from_pretrained("ctheodoris/Geneformer")
```

```python
from bertopic import BERTopic

# Create BERTopic model
topic_model = BERTopic()

# Fit the model on the documents
topics, probs = topic_model.fit_transform(documents)

# Get topic information
topic_info = topic_model.get_topic_info()

# Display the topics
print(topic_info)

# Visualize the topics
topic_model.visualize_topics()
```

```python
print (topic_info.shape)
```

```python
# Check for the presence of significant topics before visualization
if topic_info['Count'].iloc[0] > 1:  # Ensure there's more than just the default 'outliers' topic
    try:
        # Visualize the topics
        topic_model.visualize_topics()
    except ValueError as e:
        print(f"Error visualizing topics: {e}")
else:
    print("No significant topics were found.")
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Ensure there are probabilities before attempting t-SNE
if len(probs) > 0:
    # Check if probs is 1D and needs reshaping
    if np.ndim(probs) == 1:
        probs = probs.reshape(-1, 1)  # Reshape to 2D if needed
    print (probs.shape)
    # Ensure probs has more than one sample and one feature
    if probs.shape[0] > 1 and probs.shape[1] > 1:
        try:
            # Project the topic probabilities into 2D space
            tsne_model = TSNE(n_components=2, random_state=42)
            tsne_vectors = tsne_model.fit_transform(probs)

            # Plot the results
            plt.figure(figsize=(10, 6))
            scatter = plt.scatter(tsne_vectors[:, 0], tsne_vectors[:, 1], c=topics, cmap='viridis', s=5)
            plt.colorbar(scatter)
            plt.title("t-SNE projection of BERTopic topics")
            plt.xlabel("t-SNE dimension 1")
            plt.ylabel("t-SNE dimension 2")
            plt.show()
        except ValueError as e:
            print(f"Error during t-SNE visualization: {e}")
    else:
        print("Not enough samples or features for t-SNE visualization.")
else:
    print("No probabilities available for t-SNE visualization or invalid probabilities shape.")
```
